```python
import re
```

```python
def tr_lower(text):
    text = text.replace('İ', 'i').replace('I', 'ı')
    return text.lower()
```

```python
# Load the text file
with open('raw/ad_soyad.txt', 'r', encoding='utf-8') as file:
    text = tr_lower(file.read())
text[:5000]
```

```python
# Handle the Turkish I -> ı
text_split = re.split(r'\n', text)
text_split
```

```python
list_names = []
for name in text_split:
    split_name = re.split(r'\s+', name.rstrip())
    list_names.append(split_name[0])
```

```python
unique_names = list(set(list_names))
print(len(unique_names))
```
