```python
import re
from helpers import tr_lower
```

```python
with open('data/turkce_isim_veri_seti/tr_isimler.txt', 'r', encoding='utf-8') as file:
    text = file.read()
text = tr_lower(text)
text = re.split(r'\n', text)
text
```

```python
new_names = []
for name in text:
    name_split = re.split(r'\s', name)
    for subname in name_split:
        print(subname)
        if subname:
            new_names.append(subname.strip())
new_names
```

```python
new_names = list(set(new_names))
len(new_names)
```

```python
# Save the new names
with open('data/turkce_isim_veri_seti/tr_isimler2.txt', 'w', encoding='utf-8') as file:
    for line in new_names:
        file.write(line + '\n')
```
